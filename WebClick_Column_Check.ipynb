```python
# Assuming df is your DataFrame with 39 columns
results = {}
for column in train.columns:
    if column != 'Click':
      value_counts = train[column].value_counts()
      results[column] = len(value_counts[value_counts > 100000])

# results now contains the count of unique values with more than 100,000 occurrences in each column
print(results)
```

```python
import pandas as pd

# Assuming df is your DataFrame
results = {}
total_count = 0

for column in train.columns:
    value_counts = train[column].value_counts()
    frequent_values = value_counts[value_counts > 100000]
    results[column] = frequent_values.sum()  # This adds the counts of all values that appear more than 100000 times
    total_count += frequent_values.sum()

# results now contains the sum of occurrences of values with more than 100,000 occurrences in each column
print("Counts per column:", results)
print("Total count across all columns:", total_count)
```

```python
import pandas as pd

# Assuming df is your DataFrame
results = {}
total_count = 0

for column in train.columns:
    value_counts = train[column].value_counts()
    frequent_values = value_counts[value_counts > 15000]
    results[column] = frequent_values.sum()  # This adds the counts of all values that appear more than 100000 times
    total_count += frequent_values.sum()

# results now contains the sum of occurrences of values with more than 100,000 occurrences in each column
print("Counts per column:", results)
print("Total count across all columns:", total_count)
```

```python
# Assuming df is your DataFrame
for column in train.columns:
    # Calculate the value counts for each column
    value_counts = train[column].value_counts()

    # Filter to find values that appear less than 100,000 times
    to_replace = value_counts[value_counts < 100000].index

    # Replace these values with NaN
    train.loc[train[column].isin(to_replace), column] = pd.NA
```

```python
# Assuming df is your DataFrame
for column in test.columns:
    # Calculate the value counts for each column
    value_counts = test[column].value_counts()

    # Filter to find values that appear less than 15,000 times
    to_replace = value_counts[value_counts < 15000].index

    # Replace these values with NaN
    test.loc[test[column].isin(to_replace), column] = pd.NA
```

```python
import numpy as np

# Function to fill NaNs based on existing distribution
def fill_na_with_distribution(column):
    counts = column.dropna().value_counts(normalize=True)
    nans_to_fill = column.isna().sum()
    fill_values = np.random.choice(counts.index, size=nans_to_fill, p=counts.values)
    column.loc[column.isna()] = fill_values
    return column

# Apply the function to each column in DataFrame
# train = train.apply(fill_na_with_distribution, axis=0)
test = test.apply(fill_na_with_distribution, axis=0)
```
